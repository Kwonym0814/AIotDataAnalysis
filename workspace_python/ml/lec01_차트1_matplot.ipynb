```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings(action='ignore')

#-------------------- 차트 관련 속성 (한글처리, 그리드) -----------
plt.rcParams['font.family']= 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

# 기초

<pre>
* xlabel(), ylabel()  : 축 라벨
* plt.legend() : 범례
* plt.grid() : 격자
* plt.title("차트제목")           
* plt.hlines() , plt.vlines()  :  수평선/수직선
* plt.savefig("datas/test.png") : 차트 이미지 저장      

```python
plt.figure(figsize=(6, 2))                                  #-----------------------  크기
x = [1, 2, 3, 4]
y = [2, 3, 5, 10]
y1 = [6, 7, 10, 8]

plt.plot(x, y, 'ro--',   x, y1, 'bs-',  label='Price')      #-----------------------  차트   
plt.xlabel('X-Label')                                       #-----------------------  축라벨
plt.ylabel('Y-Label')

plt.fill_between(x[1:3], y1[1:3], alpha=0.5)                #-----------------------  채우기  

plt.hlines(5, 1.0, 4.0,  color='gray', linewidth=3)          #-----------------------  수평선
plt.vlines(3.5, 0, 10.0, color='blue', linewidth=2)          #-----------------------  수직선



plt.title("차트제목")                                       #-----------------------  제목
plt.grid()                                                  #-----------------------  격자
plt.legend()                                                #-----------------------  범례 (label='Price')
plt.savefig("datas/test.png")                               #-----------------------  이미지저장
plt.show()
```

## 산점도 : plt.scatter(x,y)

```python
x = np.random.rand(20)
y = np.random.rand(20)
plt.figure(figsize=(4, 2))

plt.scatter(x, y)              #---------------------------- matplot 산점도

plt.show()
```

## 파이 차트 : plt.pie(x)

```python
x = [34, 32, 16, 18]
y = ['Apple', 'Banana', 'Melon', 'Grapes']
plt.figure(figsize=(4, 2))

plt.pie(x, labels=y, autopct='%.1f%%')    #---------------------------- matplot 파이차트

plt.show()
```

## 히트맵 : plt.matshow() XXXX
* 비추

```python
# plt.figure(figsize=(4, 2))  XXXXXXXXXXXXXXXXXXXXXXXXXXXX

arr = np.random.standard_normal((30, 40))
fig, ax = plt.subplots(figsize=(1, 1))     # 원하는 크기로 Figure와 Axes 생성

mychart = ax.matshow(arr)                  # ------------------------ ax를 통해 matshow 호출
fig.colorbar(mychart)                      # fig 기준으로 colorbar 추가

plt.show()
```

# 이중 Y축 : ax2 = ax1.twinx()

```python
x = [1, 2, 3, 4]
y = [2, 3, 5, 10]
y1 = [0.2,0.8,0.5,0.4]

#plt.figure(figsize=(4, 2))   XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX

fig, ax1 = plt.subplots(figsize=(6, 2))

line1, = ax1.plot(x, y, color='green', label='범례1')
ax1.set_ylabel('Y축라벨', color='green')

ax2 = ax1.twinx()
line2, = ax2.plot(x, y1, color='deeppink', label='범례2')
ax2.set_ylabel('Y1축라벨', color='deeppink')

# 범례 표시 (핸들만 간단히 모아서 표시)
plt.legend([line1, line2], ['범례1', '범례2'])

plt.show()
```

# 멀티 차트 : plt.subplot()

<img src="https://wikidocs.net/images/page/141537/subplot_03.png" width=600>

* <b>방법1) plt.subplot(줄,칸, 1번째)

```python
plt.figure(figsize=(6, 2))

plt.subplot(1,2,1)   #----------------한줄두칸 , 1번째
x = [1, 2, 3, 4]
y = [2, 3, 5, 10]
plt.plot(x, y)
plt.title("라인차트")

plt.subplot(1,2,2)   #----------------한줄두칸 , 2번째
x = [1, 2, 3, 4]
y = [2, 3, 50, 100]
plt.bar(x, y)   
plt.title("바차트")

plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(6, 2))

axes[0][0].bar(x,y)
axes[0][0].set_title("0차트")

axes[0][1].scatter(x,y)
axes[0][1].set_title("1차트")

axes[1][0].plot(x,y)
axes[1][0].set_title("2차트")

plt.tight_layout()                 #--------------겹쳐지지 않게 그리기
plt.show()   
```

* <b>방법2)  fig, ax = plt.subplot<font size=5 color=red>s</font>(줄,칸)

```python
fig, axes = plt.subplots(2,2, figsize=(6, 2))
axes = axes.flatten()

for (i,ax) in enumerate(axes) :    #[<Axes: > <Axes: > <Axes: > <Axes: >]
    ax.plot(x,y)
    ax.set_title(f"{i}차트")

plt.tight_layout()                 #--------------겹쳐지지 않게 그리기
plt.show()    
```
