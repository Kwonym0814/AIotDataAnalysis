```python
a = ['aa.txt', 'aa.jpg', 'a100.txt']
b = []
for i in a:
    p = re.finditer('.+\.([a-z]{3})', i)
    b.extend(j.group() for j in p)
b
```

```python
a = ['aa.txt', 'aa.jpg', 'a100.txt']
for i in a:
    print(re.findall('.+\.([a-z]{3})', i))
```

```python
a = ['aa.txt', 'aa.jpg', 'a100.txt']
for i in a:
    print(re.search('.+\.(txt)', i))
```

```python
a = '02-123-1234 032-5151-1234 032-1234-5454 031-1234-5454'
p = re.compile('(\d{2,3})-\d{3,4}-\d{4}')
res = p.findall(a)
res
```

```python
p = re.compile("[ㄱ-ㅎㅏ-ㅣ가-힣]+")
res = p.findall("The name is Jack, kim, 홍길동, king, 김길동 ㅋㅋㅋ ㅠㅠㅠㅠ")
res
```
