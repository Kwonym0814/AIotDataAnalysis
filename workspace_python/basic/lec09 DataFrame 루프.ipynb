```python
import pandas as pd
import numpy as np

df = pd.read_csv('./data/dept.csv')
df

```

iterrows()

```python
for idx, row in df.iterrows():
    print(idx, row)
    print('--'*20)
```

itertuples()

```python
for tuples in df.itertuples():
    print(tuples)
    print('--'*20)
```

```python
for tuples in df.itertuples():
    print(tuples.DNAME)  # 요소만 뽑기
    print('--'*20)
```

enumerate()

```python
for idx, rows in enumerate(df):
    print(idx, rows)
    print('--'*20)
```
