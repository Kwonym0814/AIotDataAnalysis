## 딕셔너리 {}

### 조회

```python
mydic = {"name":"홍길동" ,  "age":20}
print( mydic )
```

```python
mydic["name"]
```

### 입력
* 없는 키값 넣으면 입력

```python
mydic["tel"] = "010"
mydic
```

### 수정
* 있는 키값 넣으면 수정

```python
mydic["name"] = "김길동"
mydic
```
