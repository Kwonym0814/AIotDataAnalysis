```python
sum = 0
for num in range(1,10+1) :
    if num % 2 == 0 :
        sum += num
    else :
        pass
print(sum)
```

```python
sum = 0
cumsum = {}
for num in range(1,10+1) :
    if num % 2 == 0 :
        sum += num
        cumsum[num] = sum
    else :
        pass
print(sum)
print(cumsum)
```

```python
even_sum = 0
odd_sum = 0
cum_even_sum = {}
cum_odd_sum = {}
for num in range(1,10+1) :
    if num % 2 == 0 :
        even_sum += num
        cum_even_sum[num] = even_sum
    else :
        odd_sum += num
        cum_odd_sum[num] = odd_sum
print(even_sum, odd_sum)
print(cum_even_sum)
print(cum_odd_sum)
```
