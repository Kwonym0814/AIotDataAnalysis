```python
def calc(s, e):
    tmp = 0
    for i in range(s,e+1):
        tmp+=i
    return tmp

```

```python
calc(5, 1), calc(4, 60)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(  { "name2":['kim','lee','hong','lee','lee']
                     , "age" : [10,20,np.nan,30,20]} )
df
df = pd.DataFrame(  { 
                        "grade": ['A','A','A','B','B']
                      , "age" : [10,20,np.nan,30,20]      
                      , "score" : [100,90,80,70,80]        
                      , "name2": ['kim','lee','hong','park','jung']
                  } )
df
```

```python
dfcp = df.copy()
```

```python
dfcp.groupby('grade')['age'].mean().to_dict()
```

```python
dfcp['age'] = dfcp['age'].fillna(dfcp['grade'].map(dfcp.groupby('grade')['age'].mean().to_dict()))
```
